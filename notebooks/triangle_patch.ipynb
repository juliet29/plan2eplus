```python
import shapely
from matplotlib.lines import Line2D
from replan2eplus.geometry.coords import Coord
from replan2eplus.visuals.transformations import split_coords
from replan2eplus.visuals.arrow import create_triangle_patch, line_to_coords
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
```

```python
coord_a = Coord(1,1)
coord_b = Coord(3,3)
coord_c = Coord(4,4)
line_ = Line2D(*split_coords([coord_a, coord_b, coord_c]))

line_
```

```python
EXPANSION_FACTOR=3
coords = line_to_coords(line_, -1)
line = shapely.LineString([i.as_tuple for i in coords])


dist_along = (line.length / 2) * EXPANSION_FACTOR
p0 = line.interpolate(-dist_along)


actual_dist_along = line.centroid.distance(p0)

l2 = line.parallel_offset(actual_dist_along, "right")
l3 = line.parallel_offset(actual_dist_along, "left")

poly = shapely.Polygon([p0, l2.centroid, l3.centroid])
shapely.GeometryCollection([line, p0, line.centroid, l2, l2.centroid, l3, l3.centroid, poly])
```

```python
poly = create_triangle_patch(line_, 1, 1.2)
fig, axes = plt.subplots(figsize=(12, 8))

axes.set_xlim(0,10)
axes.set_ylim(0,10)
axes.add_line(line_)
axes.add_patch(poly)
```
